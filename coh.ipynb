```python
import pandas as pd 
import numpy as np 
import scipy
from scipy.signal import *
from matplotlib import pyplot as plt
```

```python
d = pd.read_csv("sig.csv")
```

```python
fx = np.array(d["fx"])
tz = np.array(d["tz"])
pos = np.array(d["pos"])
fs = 10000
```

```python
f,co = coherence(pos,fx,fs=fs)
```

```python
plt.semilogy(f,co)
plt.xlabel("freq")
plt.ylabel("Coherence")
plt.show()
```
